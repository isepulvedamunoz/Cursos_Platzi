```python
import requests
from bs4 import BeautifulSoup
```

```python
link = 'https://www.pagina12.com.ar/secciones/deportes'
p12 = requests.get(link)
```

```python
p12.status_code
```

```python
sports = BeautifulSoup(p12.text, 'lxml')
```

```python
def noticias(soup):
    

    """ Funcion que devuelve los links de las noticias deportivas de un diario argentino,
        mediante BeautifulSoup"""
    
    """
        First, verify if the code of requests of page is 200, or return a message of error
    """
    if p12.status_code == 200:
        news = soup.find_all('div', attrs={'class':'article-item__content'})

        links_news = [new.a.get('href') for new in news]

        return links_news
    else:
        message = 'error en la pagina'
        return message

```

```python
url_notas = noticias(sports)
url_notas
```

```python
url_nota = url_notas[0]
url_nota
```

```python
def obtener_info(soup_nota):
    
    # it make a empty dictionary to append with the information
    
    ret_dict = {}
    
    # Extract date
    
    date = soup_nota.find('span', attrs={'pubdate':'pubdate'})
    if date:
        ret_dict['date'] = date.get('datetime')
    else:
        ret_dict['date'] = None
    
    #Extract title
    
    title = soup_nota.find('h1', attrs={'class':'article-title'})
    if title:
        ret_dict['title'] = title.text
    else:
        ret_dict['title'] = None
    
    #Extract prefix
    
    prefix = soup_nota.find('h2', attrs={'class':'article-prefix'})
    if prefix:
        ret_dict['prefix'] = prefix.get_text()
    
    else:
        ret_dict['prefix'] = None
    
    # Extract image
    
    media = soup_nota.find('div', attrs={'class':'article-main-media-image'})
    if media:
        imagenes = media.find_all('img')
        if len(imagenes) == 0:
            print('Not find images')
        else:
            imagen = imagenes[-1]
            img_src = imagen.get('data-src')
            try:
                img_req = requests.get(img_src)
                if img_req.status_code == 200:
                    ret_dict['imagen'] = img_req.content
                else:
                    ret_dict['imagen'] = None
            except:
                print('Not find media file')
                
    else:
        print('Not find media file')
    
    
    return ret_dict
```

```python
def scrape_nota(url):
    try:
        nota = requests.get(url)
    
    except Exception as e:
        print('Error scrapenado URL', url)
        print(e)
        return None
    
    if nota.status_code != 200:
        print(f'Error obteniendo nota {url}')
        print(f'status code = {nota.status_code}')
        return None

    s_nota = BeautifulSoup(nota.text, 'lxml')
    
    ret_dict = obtener_info(s_nota)
    ret_dict['url'] = url
    
    return ret_dict
              
```

```python
def obtener_secciones(url_secciones):
    
    secciones = requests.get(url_secciones)
    
    if secciones.status_code == 200:
        sp = BeautifulSoup(secciones.text, 'lxml')
        sections = sp.find('ul', attrs={'class':'horizontal-list main-sections'}).find_all('li')
        link_sections = [section.a.get('href') for section in sections]
    return link_sections
    
```

```python
link_section = 'https://www.pagina12.com.ar/'
url_sections = requests.get(link_section)
url_sections.status_code
```

```python
soup_sections = BeautifulSoup(url_sections.text, 'lxml')
secciones = soup_sections.find('ul', attrs={'class':'main-sections'}).find_all('li')
link_secciones = [seccion.a.get('href') for seccion in secciones]
link_secciones
```

```python
notas = []

for link in link_secciones:
    try:
        r = requests.get(link)
        if r.status_code == 200:
            soup = BeautifulSoup(r.text, 'lxml')
            notas.extend(noticias(soup))
        else:
            print('No se pudo obtener la seccion', link)
    
    except:
        print('No se pudo obtener la seccion', link)
```

```python
notas
```

```python
data = []
for i, nota in enumerate(notas):
    print(f'Scrapeando nota{i}/{len(notas)}')
    data.append(scrape_nota(nota))
```
