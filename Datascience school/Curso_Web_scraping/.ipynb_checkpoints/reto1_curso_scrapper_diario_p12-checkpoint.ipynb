```python
import requests
from bs4 import BeautifulSoup
```

```python
link = 'https://pagina12.com.ar/secciones/deportes'
p12 = requests.get(link)
```

```python
sports = BeautifulSoup(p12.text, 'lxml')
```

```python

def noticias(soup):
    

    """ Funcion que devuelve los links de las noticias deportivas de un diario argentino,
        mediante BeautifulSoup"""
    
    """
        First, verify if the code of requests of page is 200, or return a message of error
    """
    if p12.status_code == 200:
        news = soup.find_all('div', attrs={'class':'article-item__content'})

        links_news = [new.a.get('href') for new in news]

        return links_news
    else:
        message = 'error en la pagina'
        return message


    
    
```

```python
lista_links = noticias(sports)
lista_links
```

```python
url_link1 = lista_links[0]
```

```python
url_link1
```

```python
try:
    nota = requests.get(url_link1)
    if nota.status_code == 200:
        soup_nota = BeautifulSoup(nota.text, 'lxml')
        
        # extraccion del titulo de la noticia
        
        title = soup_nota.find('h1', attrs={'class':'article-title'})
        print(title.text)
        
        # extraccion de la fecha
        
        date = soup_nota.find('span', attrs={'pubdate':'pubdate'}).get('datetime')
        print(date)
        
        #extraccion de la volanta
        
        volanta = soup_nota.find('h2', attrs={'class':'article-prefix'})
        print(volanta.text)
        
        #extraccion de la imagen
        
        imagen = soup_nota.find('img', attrs={'class':'show-for-xlarge'}).get('alt')
        print(f'titulo de imagen: {imagen}')
        
        
        
        
except Exception as e:
    print('Error: ')
    print(e)
    print('\n')
```

```python
media = soup_nota.find('div', attrs={'class':'article-main-media-image'})
```

```python
imagenes = media.find_all('img')
```

```python
if len(imagenes) == 0:
    print('No hay imagenes')
else:
    imagen = imagenes[-1]
    img_src = imagen.get('data-src')
    print(img_src)
```

```python
img_req = requests.get(img_src)
```
